```python
# Install necessary libraries
!pip install chromadb pypdf langchain openai tiktoken

import chromadb
import os
import tempfile
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.text_splitter import CharacterTextSplitter
from langchain.document_loaders import PyPDFLoader
from google.colab import files

# Upload PDF
uploaded = files.upload()
file_path = list(uploaded.keys())[0]

# Load and extract text from the PDF
loader = PyPDFLoader(file_path)
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50)
splits = text_splitter.split_documents(documents)

# Initialize ChromaDB
chroma_client = chromadb.PersistentClient(path="./chroma_db")
vectorstore = Chroma.from_documents(splits, OpenAIEmbeddings(), client=chroma_client)

print("PDF processed and stored in ChromaDB successfully!")

# Function to retrieve relevant context from ChromaDB
def retrieve_context(query):
    results = vectorstore.similarity_search(query, k=3)
    return "\n".join([doc.page_content for doc in results])

# Example usage
query = "What is the document about?"
context = retrieve_context(query)
print("Retrieved Context:", context)
```
